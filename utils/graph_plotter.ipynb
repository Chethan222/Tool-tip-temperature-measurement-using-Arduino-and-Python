```python
import openpyxl as xl
from openpyxl.chart import BarChart,Reference
import matplotlib.pyplot as plt
import matplotlib.dates as dt
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import datetime
from matplotlib.ticker import MultipleLocator
from scipy.interpolate import make_interp_spline,interp1d
from bokeh.plotting import figure
from bokeh.io import output_file,show
from bokeh.io import export_png
from selenium import webdriver
from bokeh.io import curdoc
from bokeh.models import CustomJS,Slider
from bokeh.models import ActionTool
from bokeh.util.compiler import TypeScript
from bokeh.core.properties import String
from bokeh.models import Title
import datetime

MIN_TEMPERATURE = 20

def read_data(path):  
    parse_date = lambda dates: [datetime.datetime.strptime(d, '%M:%S') for d in dates]
    data = pd.read_csv(path,encoding= 'unicode_escape',parse_dates=['time(s)'],date_parser=parse_date)
    return data

def plot(data):
    Y_LIM = (MIN_TEMPERATURE,data['temperature(°C)'].max()+15)
    X_LIM = int((data['time(s)'].max() - data['time(s)'].min()).total_seconds())
    
    
    #Creating figure object
    fig = figure(plot_width=500,plot_height=400,sizing_mode='stretch_both',x_axis_type="datetime")

    label = 'Speed = '+ str(data['Speed(RPM)'][0])+'RPM'
    
    #Plotting data
    fig.line(data['time(s)'],data['temperature(°C)'],legend_label=label,color='Orange',line_width=2)
    
    #Configuring plot params
    fig.title.text = 'Temperature vs Time'
    fig.title.text_color = 'Green'
    fig.title.text_font_style = 'bold'
    fig.title.text_font = 'times'
    fig.title.text_font_size = '14pt'
    fig.title.align = 'center'

    fig.xaxis.axis_label = 'time (sec)'
    fig.yaxis.axis_label = 'Temperature (°C)'
    fig.legend.location = "bottom_right"
    fig.xaxis.minor_tick_line_color = None
    fig.yaxis.minor_tick_line_color = None
    fig.xgrid.grid_line_color = None
    fig.ygrid.grid_line_color = None

    output_file('temp_vs_time_1.html')
    
    #Adding figure caption
    figCaption = "Fig 1.3 Temperature vs Time for feed of {}mm/min and depth of cut {}mm".format(data['Feed(mm/min)'][0],data['Depth of Cut (mm)'][0])
    fig.add_layout(Title(text=figCaption, text_font = 'Times-Roman',text_font_size='16px',text_font_style = 'italic', align="center"), 'below')

    show(fig)
```

```python
path = "../data/tool-tip-temperatures_1.csv"
data = read_data(path)
plot(data)
```
